```python
# Находим строки соответсвующие шаблону
pattern = r'\d+.'
data['transId'] =data['transactionId'].str.findall(pattern).str[0]
data.head()
```

```python
# Создаем новый признак как сумму 2-х имеющихся: source и medium
data['source_medium'] = data['source'] + '_' + data['medium']
```

```python
# Оставляем только необходимые для текущего расчета колонки
df = data[['visitId', 'fullVisitorId', 'visit_time', 'transId', 'source_medium']]
```

```python
# Формируем отдельный признак, который будет складываться из visitId и fullVisitorId
df['unique_user_id'] = df['visitId'].astype(str) + df['fullVisitorId'].astype(str)
df
```

```python
# Для тестирования
# df.query('fullVisitorId == "2122547835443026286"')
```

```python
# Сортируем датасет по fullVisitorId и visit_time. visit_time был сформирован в bigquery по visitId
df = df.sort_values(by=['fullVisitorId', 'visit_time']) #.set_index(['fullVisitorId', 'visit_time'])
df.head(20)
```

```python
# Создание нового признака user_sessions. Он будет отображать количество сессий пользователя, т.е заходов в разное время
# По нему потом будет легче проверить правильность работы алгоритма
df = df.assign(
    user_sessions=df
    .groupby('fullVisitorId', group_keys=False)
    .apply(lambda x: x.groupby('visitId').ngroup()) + 1)
```

```python
# Создаем новый признак, который будет заполнять значение transId для всех вышестоящих строк имеющих Nan. Т.о сможем понять с какой транзакцией связаны строки.
# Например, имеется строка 494368, в ней была осуществлена транзакция 201707132683, следовательно все вышестоящие транзакции в рамках пользователя тоже будут относиться к ней. До следуещей транзакции.
# Если транзакции не было, то значение останется Nan
df['bfill_transact']= df.groupby(['fullVisitorId'])['transId'].bfill()
```

# Тестирование на fullVisitorId == "7701613595320832147. 
1. Удалем дубли. 
2. Группируемся по bfill_transact 
3. Применяем join по разделителю, чтобы все значения группы поместить в один столбец
4. Переименовываем колонки

```python
# Разделитель
delim = '=>'
```

```python
df.query('fullVisitorId == "7701613595320832147"')[300:]
```

```python
df_lst = (
    df
        .query('fullVisitorId == "7701613595320832147"')
        .drop_duplicates(['unique_user_id'], keep='last')
        .groupby('bfill_transact')['source_medium']
        .agg(lambda x: delim.join(x))
        .reset_index()
        .rename(columns={'bfill_transact': 'trascact_id', 'source_medium':'touchpoint'})
)
df_lst
```

```python
# Тестирование на пользователе без транзакций
df_nan = (
        df
        .query('fullVisitorId == "249021545708560"')
        .drop_duplicates(['unique_user_id'], keep='last')
        .groupby('bfill_transact')['source_medium']
        .agg(lambda x: delim.join(x))
        .reset_index()
        .rename({'bfill_transact': 'trascact_id', 'source_medium':'touchpoint'})
       
)
df_nan
```

```python
# Итоговый dataframe

df_res = (
        df
        .drop_duplicates(['unique_user_id'], keep='last')
        .groupby('bfill_transact')['source_medium']
        .agg(lambda x: delim.join(x))
        .reset_index()
        .rename(columns={'bfill_transact': 'trascact_id', 'source_medium':'touchpoint'})     
)
df_res
```

```python
# Top-10 наиболее встречающихся последовательностей касаний
top10_touchpoints = df_res['touchpoint'].value_counts().to_frame().head(10)
top10_touchpoints
```
